```python
import matplotlib.pyplot as plt
import numpy as np

# Parse the data
with open("mem.txt") as f:
    data = f.readlines()
data = data[1:]  # Skip the first line
raw_data = []
for line in data:
    data_point = line.strip().split(",")
    data_point[2] = int(data_point[2]) if data_point[2] != "incorrect" else data_point[2]
    raw_data.append(data_point)
print(raw_data)
```

```python
# extract the data
baseline_list = [data_point[2] for data_point in raw_data if data_point[1] == "baseline"]
opt_list = [data_point[2] for data_point in raw_data if data_point[1] == "opt"]
memopt_list = [data_point[2] for data_point in raw_data if data_point[1] == "memopt"]
bench_list = [data_point[0] for data_point in raw_data if data_point[1] == "baseline"]

# plots
# set width of bar 
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

# selected benchmark

# Set position of bar on X axis 
br1 = np.arange(2) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

# Make the plot
plt.bar(br1, baseline_list, color ='r', width = barWidth, 
        edgecolor ='grey', label ='Baseline') 
plt.bar(br2, opt_list, color ='g', width = barWidth, 
        edgecolor ='grey', label ='OPT') 
plt.bar(br3, memopt_list, color ='b', width = barWidth, 
        edgecolor ='grey', label ='MEMOPT') 

# Adding Xticks 
plt.xlabel('Mem Benchmark', fontweight ='bold', fontsize = 15) 
plt.ylabel('#Dynamic Instructions', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(2)], 
        bench_list, rotation=45) 

plt.legend()
plt.show() 
```
